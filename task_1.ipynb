Logistic Regression vs Multinomial NaÃ¯ve Bayes

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# 1. Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Convert to DataFrame for easy viewing
df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = [target_names[i] for i in y]

print("ðŸ”¹ First 10 rows of the Iris dataset:")
print(df.head(10))

# 2. Plot distribution of the data
sns.pairplot(df, hue="species", diag_kind="kde")
plt.suptitle("Distribution of Iris Dataset", y=1.02)
plt.show()

# 3. Train-test split
scaler = MinMaxScaler()  # Needed for Naive Bayes (non-negative features)
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# 4. Logistic Regression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
y_pred_log = log_reg.predict(X_test)

# 5. Multinomial Naive Bayes
nb = MultinomialNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)

# 6. Evaluation
print("\nLogistic Regression Accuracy:", accuracy_score(y_test, y_pred_log))
print("Naive Bayes Accuracy:", accuracy_score(y_test, y_pred_nb))

print("\nLogistic Regression Report:\n", classification_report(y_test, y_pred_log, target_names=target_names))
print("\nNaive Bayes Report:\n", classification_report(y_test, y_pred_nb, target_names=target_names))

print("\nConfusion Matrix - Logistic Regression:\n", confusion_matrix(y_test, y_pred_log))
print("\nConfusion Matrix - Naive Bayes:\n", confusion_matrix(y_test, y_pred_nb))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
import pandas as pd

#  Load the Iris dataset from CSV
df = pd.read_csv("iris.csv")  # <-- Make sure 'iris.csv' is in your working directory

print(" First 10 rows of the Iris dataset:")
print(df.head(10))

#  Encode species column (convert text labels to numbers)
label_encoder = LabelEncoder()
df['species_encoded'] = label_encoder.fit_transform(df['species'])

# Separate features (X) and target (y)
X = df.drop(['species', 'species_encoded'], axis=1).values
y = df['species_encoded'].values
feature_names = df.drop(['species', 'species_encoded'], axis=1).columns
target_names = label_encoder.classes_

#  Pairplot visualization
sns.pairplot(df, vars=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],
             hue='species')
plt.suptitle("Distribution of Iris Dataset", y=1.02)
plt.show()

#  Train-test split
scaler = MinMaxScaler()  # Normalize for Naive Bayes
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

#  Logistic Regression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
y_pred_log = log_reg.predict(X_test)

# Multinomial Naive Bayes
nb = MultinomialNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)

#  Evaluation
print("\nLogistic Regression Accuracy:", accuracy_score(y_test, y_pred_log))
print("Naive Bayes Accuracy:", accuracy_score(y_test, y_pred_nb))

print("\nLogistic Regression Report:\n", classification_report(y_test, y_pred_log, target_names=target_names))
print("\nNaive Bayes Report:\n", classification_report(y_test, y_pred_nb, target_names=target_names))

print("\nConfusion Matrix - Logistic Regression:\n", confusion_matrix(y_test, y_pred_log))
print("\nConfusion Matrix - Naive Bayes:\n", confusion_matrix(y_test, y_pred_nb))

# Feature Importance (Logistic Regression)
# Take the absolute value of coefficients and average across classes
importance = np.mean(np.abs(log_reg.coef_), axis=0)

# Create a DataFrame for plotting
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importance
}).sort_values(by='Importance', ascending=True)

#  Plot Feature Importance (horizontal bar)
plt.figure(figsize=(8, 5))
plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.xlabel("Feature Importance")
plt.ylabel("Feature")
plt.title("Feature Importance (Logistic Regression - Iris Dataset)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np


# 1ï¸âƒ£ Load dataset
df = pd.read_csv("spam.csv", encoding="latin-1")

# Keep only useful columns
df = df.rename(columns={"v1": "label", "v2": "text"})
df = df[["label", "text"]]

print("ðŸ”¹ First 10 rows of the Spam dataset:")
print(df.head(10))

# 2ï¸âƒ£ Encode target variable (spam = 1, ham = 0)
label_encoder = LabelEncoder()
df["label_encoded"] = label_encoder.fit_transform(df["label"])
target_names = list(label_encoder.classes_)  # ['ham', 'spam']

# 3ï¸âƒ£ Show class distribution
sns.countplot(x="label", hue="label", data=df, palette="coolwarm", legend=False)
plt.title("Distribution of Spam vs Ham Messages")
plt.show()

# 4ï¸âƒ£ Convert text data into numeric features using TF-IDF
vectorizer = TfidfVectorizer(stop_words="english", max_features=1000)
X = vectorizer.fit_transform(df["text"])
y = df["label_encoded"].values

# 5ï¸âƒ£ Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 6ï¸âƒ£ Train Logistic Regression (just like in Iris example)
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# 7ï¸âƒ£ Evaluation
print("\nLogistic Regression Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=target_names))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

importance = np.abs(model.coef_[0])
feature_names = np.array(vectorizer.get_feature_names_out())

# Create DataFrame for visualization
importance_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance": importance
}).sort_values(by="Importance", ascending=False).head(20)  # Top 20 most important words

# 7ï¸âƒ£ Plot Feature Importance
plt.figure(figsize=(10, 6))
plt.barh(importance_df["Feature"][::-1], importance_df["Importance"][::-1])
plt.xlabel("Importance (Weight Magnitude)")
plt.ylabel("Word (TF-IDF Feature)")
plt.title("Top 20 Important Words for Spam Detection (Logistic Regression)")
plt.tight_layout()
plt.show()
```
